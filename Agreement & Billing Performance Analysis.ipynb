```python

df_balance = pd.DataFrame({
    'agr_number': ['SEPL/07','AGR002','MAZAGON','S.NO - 02','AGR280'],
    'billing_total_balance': [45000, 120000, 80000, 30000, 50000]
})
# Sort descending
df_pareto = df_balance.sort_values("billing_total_balance", ascending=False)

# Calculate cumulative percentage
df_pareto['cum_percent'] = df_pareto['billing_total_balance'].cumsum() / df_pareto['billing_total_balance'].sum() * 100

fig, ax1 = plt.subplots(figsize=(12,6))

# Bar chart for balance
ax1.bar(df_pareto['agr_number'], df_pareto['billing_total_balance'])
ax1.set_ylabel("Billing Outstanding Amount")
ax1.set_xlabel("Agreement Number")

# Line chart for cumulative %
ax2 = ax1.twinx()
ax2.plot(df_pareto['agr_number'], df_pareto['cum_percent'], marker='o', color='red')
ax2.set_ylabel("Cumulative %")

plt.title("Pareto Analysis of Outstanding Billing Balances")
plt.tight_layout()
plt.show()
```
